## Move files

```python
import shutil
import os
from pathlib import Path
import glob
    
source_dir = Path.cwd().parent / "runtime" / "data" / "train_features" / "train_features"
target_dir = Path.cwd().parent / "runtime" / "data" / "train_features" / "additional_features"
    
#for file_name in file_names:
#    shutil.move(os.path.join(source_dir, file_name), target_dir)
```

```python
file_names = os.listdir(source_dir)
correct_files = []
for file in file_names:
    if((file[-7:] != "_vv.tif") & (file[-7:] != "_vh.tif")):
        correct_files.append(file)
```

## Copy additional data to test submission folder

```python
import shutil
import os
from pathlib import Path
import glob
    
source_dir = Path.cwd().parent / "runtime" / "data" / "train_features" / "additional_features"

endings = ["-gsw-change.tif", "-gsw-extent.tif", "-occurrence.tif", "-recurrence.tif", "seasonality.tif", "transitions.tif",
"asadem.tif"]

dirs = ["jrc_change", "jrc_extent", "jrc_occurrence", "jrc_recurrence", "jrc_seasonality", "jrc_transitions", "nasadem"]
  
target_dirs = []

for i in range(7):
    target_dirs.append(Path.cwd().parent / "test submission" / "codeexecution" / "data" / dirs[i])    
```

```python
file_names = os.listdir(source_dir)

for file in file_names:
    for i in range(7):
        if(file[-10:] == endings[i]):
            shutil.copy(os.path.join(source_dir, file), target_dirs[i])
        elif(file[-15:] == endings[i]):
            shutil.copy(os.path.join(source_dir, file), target_dirs[i])
```
